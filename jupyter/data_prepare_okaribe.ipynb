```python
import pandas as pd
import numpy as np
import sys
sys.path.append('../')
sys.path.append('../utils/')
from utils import common_utils
from utils import data_prepare_utils
import seaborn as sns
from glob import glob 
```

```python
okaribe_train_path = '/data/renjie/tunnel_cv/okaribe_bk/data/train/masterdataset/'
okaribe_test_path = '/data/renjie/tunnel_cv/okaribe_bk/data/test/masterdataset/'
train_paths = glob(okaribe_train_path+'*.csv')
test_paths = glob(okaribe_test_path+'*.csv')
paths = train_paths+test_paths
paths
```

```python
dfs = []
for p in paths:
    dfs.append(pd.read_csv(p))
    
```

```python
okaribe_df = pd.concat(dfs)
okaribe_df['datetime'] = pd.DatetimeIndex(okaribe_df['datetime'])
okaribe_df['path'] = okaribe_df['name']

okaribe_df = okaribe_df.sort_values(['datetime'])
okaribe_df = okaribe_df.drop(columns = ['name'])
```

```python
okaribe_df
```

```python
full_save_path = '../data/okaribe_full.csv'
okaribe_df.to_csv(full_save_path,index=False)
```

```python
okaribe_train_df = okaribe_df[okaribe_df['datetime'] < '2019-04-01']
okaribe_test_df = okaribe_df[okaribe_df['datetime'] > '2019-04-01']
```

```python
train_save_path = '../data/train_data/okaribe_train.csv'
test_save_path = '../data/test_data/okaribe_test.csv'

okaribe_train_df.to_csv(train_save_path,index=False)
okaribe_test_df.to_csv(test_save_path,index=False)
```
